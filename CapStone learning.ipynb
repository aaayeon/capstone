
#데이터정리



```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('/content/drive/MyDrive/input_word_data_2.csv', index_col = "name")
data = data.dropna()
data.head()
```

```python
X = data.iloc[:, :2]
y = data.iloc[:, 2]
```

```python
X
```

```python
y
```

```python
X_ = pd.get_dummies(X)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_, y, test_size = 0.2, shuffle=False, random_state = 5)
```

```python
X_test
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
result = data.iloc[400:, :]
```

#KNeighborsClassifier

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=10)
```

```python
knn.fit(X_train, y_train)
```

```python
knn_prd = knn.predict(X_test)
```

```python
knn.score(X_test, y_test)
```

```python
result_knn = result.assign(result=knn_prd)
result_knn
```

#분류용 선형 모델

```python
from sklearn.linear_model import LogisticRegression
```

```python
logreg = LogisticRegression()
```

```python
logreg.fit(X_train, y_train)
```

```python
logreg_prd = logreg.predict(X_test)
```

```python
logreg.score(X_train, y_train) # 훈련 세트 점수
```

```python
logreg.score(X_test, y_test) # 테스트 세트 점수
```

```python
result_logreg = result.assign(result=logreg_prd)
result_logreg
```

# 모델 저장
 - 학습한 모델을 pickle 형태로 변수에 저장한 것

```python
import pickle
from sklearn.externals import joblib
```

```python
saved_model_gbrt = pickle.dumps(gbrt)
```

```python
# Load the pickled model
gbrt_from_pickle = pickle.loads(saved_model_gbrt)

# Use the loaded pickled model to make predictions
gbrt_from_pickle.predict(X_test)
```

 - 파일에 저장

```python
joblib.dump(gbrt, 'english_gbrt.pkl') 
```
